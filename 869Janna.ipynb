```python
#Import Files
Train_label = pd.read_csv('Training set labels.csv')
Train_value = pd.read_csv('Training set values.csv')
Test_value = pd.read_csv('Test set values.csv')

```

```python
# Merger Train value and label 
Train = pd.merge(Train_value,Train_label, how='inner',left_on='id',right_on='id', left_index=False)
```

```python
# For Data Explore
Train.to_csv('Train.csv')
Train.head()
Train.shape
Train.tail()
Train.isnull().sum()
```

```python
#source_water = Train[["source","source_class","source_type","waterpoint_type","waterpoint_type_group", "status_group"]]
source_water = Train.iloc[:,35:41]
source_water.head()

#df_no_label[df_no_label.select_dtypes(['object']).columns] = df_no_label.select_dtypes(['object']).apply(lambda x: x.astype('category'))
source_water[source_water.select_dtypes(['object']).columns] = source_water.select_dtypes(['object']).apply(lambda x: x.astype('category'))
source_water.dtypes
```

```python
#Label Ordinal Encoding
source_water['source_cat'] = source_water['source'].cat.codes
source_water['source_class_cat'] = source_water['source_class'].cat.codes
source_water['source_type_cat'] = source_water['source_type'].cat.codes
source_water['waterpoint_type_cat'] = source_water['waterpoint_type'].cat.codes
source_water['waterpoint_type_group_cat'] = source_water['waterpoint_type_group'].cat.codes
source_water['status_group_cat'] = source_water['status_group'].cat.codes
source_water.head()
```
